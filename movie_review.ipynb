```python
useless_words = nltk.corpus.stopwords.words("english")+list(string.punctuation)
useless_words
```

```python
def build_bag_of_words_features_filtered(words):
    return{
        word:1 for word in words \
        if not word in useless_words}
```

```python
build_bag_of_words_features_filtered(romeo_words)
```

```python
negative_features = [
    (build_bag_of_words_features_filtered
    (movie_reviews.words(fileids=[f])),'neg')\
    for f in negative_fields
]

```

```python
negative_fields
```

```python
positive_features = [
    (build_bag_of_words_features_filtered
    (movie_reviews.words(fileids=[f])),'pos')\
    for f in positive_fields
]
```

```python
positive_fields
```

```python
from nltk.classify import NaiveBayesClassifier
```

```python
split = 800
```

```python
sentiment_classifier = NaiveBayesClassifier.train(positive_features[:split]+negative_features[:split])
```

```python
nltk.classify.accuracy(sentiment_classifier,positive_features[:split]+negative_features[:split])*100
```

```python
nltk.classify.accuracy(sentiment_classifier,positive_features[split:]+negative_features[split:])*100
```

```python
sentiment_classifier.show_most_informative_features()
```
